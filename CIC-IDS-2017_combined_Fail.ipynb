```python
def cap_outliers_iqr_chunkwise(df, chunk_size=50000, use_train_bounds=False, train_bounds=None):
    """
    Applies IQR-based outlier capping in chunks.
    If `use_train_bounds` is True, uses precomputed `train_bounds` to cap test data.
    Returns capped DataFrame and train_bounds if applicable.
    """
    df_out = []
    bounds = {} if not use_train_bounds else train_bounds

    n = df.shape[0]
    for start in range(0, n, chunk_size):
        end = min(start + chunk_size, n)
        chunk = df.iloc[start:end].copy()

        for col in chunk.select_dtypes(include=[np.number]).columns:
            if not use_train_bounds:
                Q1 = chunk[col].quantile(0.25)
                Q3 = chunk[col].quantile(0.75)
                IQR = Q3 - Q1
                if IQR == 0 or pd.isna(IQR):
                    continue
                low, high = Q1 - 1.5 * IQR, Q3 + 1.5 * IQR
                bounds[col] = (low, high)
            else:
                if col not in bounds:
                    continue
                low, high = bounds[col]

            chunk[col] = chunk[col].clip(lower=low, upper=high)

        df_out.append(chunk)

    return (pd.concat(df_out, ignore_index=True), bounds) if not use_train_bounds else pd.concat(df_out, ignore_index=True)
```

```python
from sklearn.preprocessing import PowerTransformer, StandardScaler

def transform_chunkwise_yeojohnson(df, columns, chunk_size=50000):
    """
    Applies Yeo-Johnson transformation + Standardization chunkwise.
    Returns transformed DataFrame and transformers for reuse.
    """
    transformers = {}
    df_out = []

    # Fit on the first chunk to establish parameters
    first_chunk = df.iloc[:chunk_size].copy()
    transformed_chunk = first_chunk.copy()

    for col in columns:
        pt = PowerTransformer(method='yeo-johnson', standardize=False)
        scaler = StandardScaler()

        values = first_chunk[[col]].values
        values = pt.fit_transform(values)
        values = scaler.fit_transform(values)

        transformers[col] = (pt, scaler)
        transformed_chunk[col] = values.flatten()

    df_out.append(transformed_chunk)

    # Transform remaining chunks
    n = df.shape[0]
    for start in range(chunk_size, n, chunk_size):
        end = min(start + chunk_size, n)
        chunk = df.iloc[start:end].copy()

        for col in columns:
            pt, scaler = transformers[col]
            values = chunk[[col]].values
            values = pt.transform(values)
            values = scaler.transform(values)
            chunk[col] = values.flatten()

        df_out.append(chunk)

    return pd.concat(df_out, ignore_index=True), transformers
```

```python
from tensorflow.keras.layers import Bidirectional, Attention

inp = Input((X_train.shape[1],))
x = Dense(128, activation='relu')(inp)
x = Dense(64, activation='relu')(x)
x = Reshape((1,64))(x)
x = Conv1D(64,1,activation='relu')(x)
x = Conv1D(32,1,activation='relu')(x)
x = Bidirectional(LSTM(32, return_sequences=True))(x)

# Attention
att = Attention()([x, x])
x = LSTM(32)(att)

x = Dropout(0.3)(x)
x = Dense(32, activation='relu')(x)
out = Dense(num_classes, activation='softmax')(x)

model = Model(inp, out)
model.compile(optimizer='adam',
              loss=tf.keras.losses.CategoricalFocalCrossentropy(),
              metrics=['accuracy'])

model.summary()
```

```python
# Optional: Early stopping
early_stop = EarlyStopping(monitor='val_loss', patience=3, restore_best_weights=True)

# Fit the model
history = model.fit(
    X_train, y_train,
    validation_data=(X_test, y_test),
    epochs=20,
    batch_size=1024,
    callbacks=[early_stop],
    verbose=1
)
```
